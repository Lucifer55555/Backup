```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('iris.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#dropping id column as it is not useful
df.drop('Id', axis=1, inplace=True)

## creating code catrgoty for species in target column
df['target']=df.Species.astype('category').cat.codes
```

```python
df.info()
```

```python
df.target.value_counts()
```

```python
#features
x=df.iloc[:,:-1]
x.drop('Species', axis=1, inplace=True)
#target
y=df.iloc[:,-1:]
```

```python
y
```

# Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
```

```python
x=scaler.fit_transform(x)
```

# Machine Learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
log_ovr=LogisticRegression(multi_class='ovr')
log_ovr.fit(x,y)
print("score on log_ovr", log_ovr.score(x,y))
```

# Let’s look at the confusion matrix and classification reports to explore a little more on the performance of our models.

```python

#predictions
pred_ovr=log_ovr.predict(x)

#confusion matrix and classification report
print('multi_class=ovr')
print(confusion_matrix(y,pred_ovr, labels=[0,1,2]))
print(classification_report(y,pred_ovr, labels=[0,1,2]))
```

```python
log_mul=LogisticRegression(multi_class='multinomial')
log_mul.fit(x,y)
print("score on log_mul", log_mul.score(x,y))

#predictions
pred_mul=log_mul.predict(x)

#confusion matrix and classification report
print('multi_class=mul')
print(confusion_matrix(y,pred_mul, labels=[0,1,2]))
print(classification_report(y,pred_mul, labels=[0,1,2]))
```
