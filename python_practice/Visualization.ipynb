# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.arange(0,11,2.5)
```

```python
x=np.linspace(0,10,100)
```

plt.plot(x, np.sin(x))

__%matplotlib inline -- it generates a static plot,this is default__
__%matplotlib notebook -- makes the plot interactive__

```python
%matplotlib notebook
plt.plot(x, np.sin(x),"-")
plt.plot(x, np.cos(x),".")
plt.plot(x, np.tan(x),"--")
plt.show()
```

### SUBPLOT

```python
plt.style.use('classic')
plt.figure(figsize=(5,7))

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()

```

```python
plt.style.use('classic')
plt.figure(figsize=(7,7))

plt.subplot(2,2,4)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, np.tan(x))


plt.show()

```

```python
fig, ax=plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.sin(x))
ax[1,1].plot(x, np.tan(x))

plt.show()
```
